```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Date': pd.date_range(start='1/1/2022', periods=7, freq='D'),
    'Value1': np.random.rand(7) * 100,
    'Value2': np.random.rand(7) * 100,
    'Value3': np.random.rand(7) * 100,
    'Value4': np.random.rand(7) * 100,
    'Value5': np.random.rand(7) * 100,
    'Value6': np.random.rand(7) * 100,
}
df = pd.DataFrame(data)

plt.figure(figsize=(11, 6))
plt.plot(df['Date'], df['Value1'], label='Red', color='red')
plt.plot(df['Date'], df['Value2'], label='Blue', color='blue')
plt.plot(df['Date'], df['Value3'], label='Yellow', color='yellow')
plt.plot(df['Date'], df['Value4'], label='Green', color='green')
plt.plot(df['Date'], df['Value5'], label='Orange', color='orange')
plt.plot(df['Date'], df['Value6'], label='Purple', color='purple')
plt.title('Sample Line Chart')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.show()
```
